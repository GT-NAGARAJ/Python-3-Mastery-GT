SECTION 3: MUTABLE â€“ NORMAL CASES

```python
skills = ["Python", "SQL"]
skills_alias = skills

skills.append("AWS")

print(skills)
print(skills_alias)
```

```python
list1 = [1, 2, 3]
list2 = list1

list2 = list2 + [4]

print(list1)
print(list2)
```

```python
list1 = [1, 2, 3]
list2 = list1

list2.append(4)

print(list1)
print(list2)

#
# ðŸ’¡ Hint:

# In-place modification
```

SECTION 4: MUTABLE â€“ FUNCTION LEVEL

```python
def add_skill(skills):
    skills.append("Docker")

emp_skills = ["Python"]
add_skill(emp_skills)
print(emp_skills)

# ðŸ§  Think:

# Same object passed

# In-place change
```

```python
def add_skill_safe(skills):
    skills = skills.copy()
    skills.append("Kubernetes")

emp_skills = ["Python"]
add_skill_safe(emp_skills)
print(emp_skills)

# ðŸ’¡ Hint:

# Copy breaks reference
```

```python
def reassign_list(skills):
    skills = ["Go", "Rust"]

emp_skills = ["Python", "SQL"]
reassign_list(emp_skills)
print(emp_skills)


# ðŸ§  Think:

# Reassignment vs mutation
```

```python
def process(data):
    data.append(100)
    data = [200, 300]

lst = [1, 2]
process(lst)
print(lst)

# ðŸ’¡ Hint:

# One mutation

# One reassignment
```

```python
def update(emp):
    emp["salary"] += 5000

employee = {"name": "Ravi", "salary": 50000}
update(employee)
print(employee)
```

```python
def update_skills(emp):
    emp["skills"].append("AWS")

employee = {
    "name": "Ravi",
    "skills": ["Python"]
}

update_skills(employee)
print(employee)
```
