| Scenario             | Best Data Type | Why              |
| -------------------- | -------------- | ---------------- |
| Employee ID          | `int`          | Whole number     |
| Salary               | `float`        | Decimal values   |
| Name / Email         | `str`          | Text             |
| Skills               | `list`         | Changeable       |
| Unique skills        | `set`          | No duplicates    |
| Fixed attributes     | `tuple`        | Immutable        |
| Full employee record | `dict`         | Structured       |
| Status flags         | `bool`         | True / False     |
| Missing data         | `None`         | Safe placeholder |
