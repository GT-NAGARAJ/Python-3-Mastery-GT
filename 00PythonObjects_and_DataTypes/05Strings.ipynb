## String Indexing  and slicing

```python
s = "Hello World I'm GT"
```

```python
s
```

```python
print(s[0])
print(s[1])
print(s[-1])
print(s[:])
print(s[13])

print(s[5:10])
print(s[::2])
print(s[3:10:3])
print(s[::-1])


```

## Strings are Immutable

```python
s[0]='T'
```

# But it can be concatinated

```python
print(id(s))
s= s+' Nagaraju'
print(s)
print(id(s))

# A new object is created when we concatenated the string as it's immutable.
```
